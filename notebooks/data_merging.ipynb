## Load data

```python
import pandas as pd

medal = pd.read_csv("../data/raw/medals_country.csv")
medal.head()
```

```python
population_wiki = pd.read_csv("../data/raw/population_wikipedia.csv")
population_wiki.head()
```

```python
population_world = pd.read_csv("../data/raw/worldometers_population.csv")
population_world.head()
```

```python
temp = pd.read_csv("../data/raw/temperature.csv")
temp.head()
```

```python
gdp = pd.read_csv("../data/raw/gdp_per_capita.csv")
gdp.head()
```

## Merge Medal and Population data

```python
df = pd.merge(medal, population_wiki, left_on='Country', right_on='country', how='left')
df
```

```python
df.loc[df['population'].isna(), 'Country']
```

```python
country_mapping = {
    'United States of America': 'United States',
    "People's Republic of China": 'China',
    'Republic of Korea': 'South Korea',
    'Great Britain': 'United Kingdom'
}

df['country'] = df['Country'].replace(country_mapping)
```

```python
df = df.drop(columns=['population'])
df = df.merge(population_wiki, on='country', how='left')
```

```python
df = df.rename(columns={'population': 'population_wiki'})
```

```python
df
```

```python
df = df.merge(population_world, on='country', how='left')
df
```

```python
df[df['population'].isna()]
```

```python
population_world[population_world['country'].str.contains("Czech")]
```

```python
population_world_mapping = {
    'Czech Republic (Czechia)': 'Czechia'
}

population_world['country'] = population_world['country'].replace(population_world_mapping)
```

```python
df = df.drop(columns=['population'], errors='ignore')
df = df.merge(population_world[['country', 'population']], on='country', how='left')
df = df.rename(columns={'population': 'population_world'})
```

```python
df
```

```python
df['population'] = (df['population_wiki'] + df['population_world'])/2
df['population'] = df['population'].astype('int')
df = df.drop(columns=['population_wiki', 'population_world'])
df
```

## Merge GDP data

```python
df = df.merge(gdp, on='country', how='left')
```

```python
df[df['gdp_per_capita_usd'].isna()]
```

```python
gdp[gdp['country'].str.contains('Korea', na=False)]
```

```python
gdp_mapping = {
    'Korea, Rep.': 'South Korea'
}

gdp['country'] = gdp['country'].replace(gdp_mapping)
```

```python
df = df.drop(columns=['gdp_per_capita_usd'])

df = df.merge(gdp[['country', 'gdp_per_capita_usd']], on='country', how='left')
```

```python
df = df.drop(columns=['country_code', 'region'])
df
```

## Merge Temperature data

```python
temp.head()
```

```python
temp_5_years = temp[temp['year'].isin([2018,2019,2020,2021,2022])]
temp_5_years.head(10)
```

```python
temp_new= temp_5_years.groupby('country', as_index=False)['temperature'].mean()
```

```python
temp_new
```

```python
df = df.merge(temp_new, on='country', how='left')
```

```python
df[df['temperature'].isna()]
```

```python
temp_new[temp_new['country'].str.contains("korea", case=False)]
```

```python
temp_new[temp_new['country'].str.contains("united", case=False)]
```

```python
temp_mapping = {
    'Korea, Rep.': 'South Korea',
    'United States of America' : 'United States'
}

temp_new['country'] = temp_new['country'].replace(temp_mapping)
```

```python
df = df.drop(columns=['temperature'])

df = df.merge(temp_new[['country', 'temperature']], on='country', how='left')
```

```python
df
```

## save dataset
